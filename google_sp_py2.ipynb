```python
import librosa
import speech_recognition as sr
import time

save_list=[]

def voice(filepath):
    voice_lenth=librosa.get_duration(filename=filepath)
    
    x=0
    i=(voice_lenth/5)
    save_list.append("此音檔總秒數"+str(voice_lenth))
    
    for part in range(int(i)):
        x=x+5
        r = sr.Recognizer()
        time.sleep(1)
        with sr.WavFile(filepath) as source:    #讀取wav檔
            audio = r.record(source,offset=x, duration=5)
        try:
            save_list.append(str(x)+"～"+str(x+5)+"sec")#使用Google的服務
            save_list.append(r.recognize_google(audio,language="zh-TW"))
        except:
            save_list.append("未讀取到人聲")
    return(save_list)
```
