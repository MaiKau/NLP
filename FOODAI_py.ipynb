```python
import requests
import json
from google_trans_new import google_translator

# filePath = "./steak.jpg"
#api地址
url = 'https://api.foodai.org/v4.1/classify'
#api申請的 key
api_key = 'a3be469153a30264e9245a2d39eadc336ff4de25'

#回傳的格式參數
def json_params(_img, _qid=-1, _num_tag=10, _api_key=''):
    return dict(
        image_url=_img,
        num_tag=_num_tag,
        qid=_qid,
        api_key=_api_key
        )

#post
def main():
    # Open Image File
    with open(filePath, 'rb') as f:
        image_bytes = f.read()
        image_data = dict(image_data=image_bytes)

    # provide your api key and add more parameters in post body if you want.
    json_data = dict(api_key=api_key)

    # Send Requests
    resp = requests.post(url=url, files=image_data, data=json_data)

    # Check the JSON Response Content documentation below
    #print(resp.json())

#辨識與字串處理
def openimg(filePath):
    with open(filePath, 'rb') as f:                #rb 二進制
        image_bytes = f.read()
        files = {'image_data': image_bytes}
    
        # provide your api key and add more  parameter in  post body if  you want.
        # or  more detail  of   parameter you can check developer  doc- https://foodai.org/developer#doc

        d = {'api_key':api_key}
        text = requests.post(url, files=files, data = d).text 
    json_str=json.loads(text)
    food_results=json_str["food_results"]
    food_results_by_category=json_str["food_results_by_category"]
    
    translate = google_translator ()
    food_results_json = json.dumps(translate.translate (json_str ,lang_tgt = 'zh-TW'), ensure_ascii=False) #避免被轉成ASCII編碼
    food_results_text = json.dumps(translate.translate (food_results ,lang_tgt = 'zh-TW'),ensure_ascii=False)
    food_results_by_category_text = json.dumps(translate.translate (food_results_by_category ,lang_tgt = 'zh-TW'),ensure_ascii=False)
    
    # 回傳辨識結果，降冪排序。
   # text:原回傳格式 ; food_results_json:原文翻譯 ; food_results_text:分割結果1 ; food_results_by_category_text:分割結果2
    return(json_str , food_results_json , food_results_text , food_results_by_category_text)
```
