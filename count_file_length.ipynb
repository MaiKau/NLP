```python
import os
import librosa
import speech_recognition as sr
import soundfile
import time
from pandas import DataFrame
import pandas as pd

timelist = []
dirpath  = r"./Tofile/"
result  = [f for f in os.listdir(dirpath) if os.path.isfile(os.path.join(dirpath,f))]
print (len(result),result)

i=-1
for lenth in result:
    i=i+1
    data, samplerate = soundfile.read(dirpath+result[i])
    soundfile.write(dirpath+result[i], data, samplerate, subtype='PCM_16')
    voice_lenth=librosa.get_duration(filename=dirpath+result[i])
    print(int(voice_lenth+1))
    timelist.append(int(voice_lenth+1))

print("--------------------------------------------------------------")
print(timelist)
```
